Kasus ini anda ditugaskan menghitung nilai perkiraan dari suatu Fungsi Evaluasi $f(x)$ (tiap mahasiswa memiliki jenis fungsi $f(x)$ yang berbeda) menggunakan Deret Tak Hingga. Akurasi perhitungan harus ditentukan oleh Batas Toleransi.

Tugas Komputasi:

Definisikan Fungsi: Buatlah sebuah fungsi di Python dengan nama spesifik sesuai kasus (misalnya, `hitung_deret_evaluasi`) yang menerima `x, N,` dan `batas_toleransi`.

Perulangan dan Deret: Gunakan Perulangan (`for` atau `while`) untuk menghitung jumlah suku dari Deret Evaluasi sesuai yang ditugaskan (lihat tabel di atas).

Kriteria Penghentian (`break`): Terapkan struktur kontrol `if` dan `break` untuk menghentikan perhitungan jika selisih nilai absolut antara Suku ke-$(n+1)$ dan Suku ke-$n$ sudah lebih kecil dari `batas_toleransi`.$$|Suku_{n+1} - Suku_n| < \text{batas\_toleransi}$$

Output: Fungsi harus mengembalikan (return) sebuah tuple yang berisi `(Nilai Hasil Deret, Jumlah Suku yang Digunakan, N Batas Maksimum)`.
