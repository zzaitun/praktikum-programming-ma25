## 1. Konsep Dasar While Loop

Perulangan (*loop*) digunakan untuk menjalankan satu blok perintah **berulang kali** selama suatu kondisi bernilai `True`.

Sintaks dasar:
```python
while kondisi:
    perintah_yang_dijalankan
```

- Program akan terus menjalankan perintah di dalam `while` **selama** kondisi bernilai `True`.
- Jika kondisi sudah tidak terpenuhi (`False`), maka perulangan berhenti.

## 2. Contoh Sederhana While Loop

Menampilkan angka 1 sampai 5 menggunakan `while`.

````python
i = 1
while i <= 5:
    print(i)
    i += 1  # menambah nilai i setiap iterasi
````

ðŸ’¡ **Penjelasan:**
- `i = 1` â†’ inisialisasi.
- `while i <= 5:` â†’ selama nilai `i` kurang dari atau sama dengan 5, lakukan perulangan.
- `i += 1` â†’ nilai `i` bertambah 1 setiap kali loop dijalankan.

## 3. While Loop dengan Kondisi dan Counter

Contoh: Menampilkan bilangan genap dari 2 hingga 10.

````python
angka = 2
while angka <= 10:
    print(angka)
    angka += 2
````

## 4. While Loop dengan Input Pengguna

Contoh: Program akan terus meminta input angka hingga pengguna memasukkan angka 0.

````python
angka = int(input("Masukkan angka (0 untuk berhenti): "))
while angka != 0:
    print("Anda memasukkan angka:", angka)
    angka = int(input("Masukkan angka (0 untuk berhenti): "))

print("Program selesai.")
````

## 5. Menggunakan `break` dan `continue`

### a) `break` â†’ menghentikan perulangan secara paksa
### b) `continue` â†’ melewati satu iterasi, lalu lanjut ke berikutnya.

````python
# Contoh penggunaan break
i = 1
while i <= 10:
    if i == 6:
        break  # hentikan loop saat i == 6
    print(i)
    i += 1
````

````python
# Contoh penggunaan continue
i = 0
while i < 5:
    i += 1
    if i == 3:
        continue  # lewati angka 3
    print(i)
````

## 7. Mini Project: Simulasi Penghitungan Nilai

Program berikut menggunakan `while` untuk menghitung rata-rata nilai mahasiswa hingga pengguna mengetik `selesai`.

````python
total = 0
jumlah = 0

while True:
    nilai = input("Masukkan nilai (ketik 'end' untuk berhenti): ")
    if nilai.lower() == 'end':
        break
    total += int(nilai)
    jumlah += 1

if jumlah > 0:
    rata = total / jumlah
    print("Rata-rata nilai =", rata)
else:
    print("Tidak ada nilai yang dimasukkan.")
````
